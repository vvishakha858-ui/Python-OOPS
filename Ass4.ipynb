#4- What is the difference between abstraction and encapsulation?

 - Both Abstraction and Encapsulation are key concepts of Object-Oriented Programming (OOP).
 They sound similar but serve different purposes.

 1. Definition

    - Abstraction → Hiding the implementation details and showing only the essential features of an object.

    -  Encapsulation → Wrapping data (variables) and methods (functions) together into a single unit (class).



---

 2. Purpose

     - Abstraction → Focuses on what an object does, not how it does it.

     - Encapsulation → Focuses on how to protect data and keep it safe from direct access.



---

3. Implementation

   - Abstraction is implemented using abstract classes and interfaces (in languages like Java, C++).
   - In Python, it can be achieved using abstract base classes (abc module).

    - Encapsulation is implemented using classes, access modifiers (private, protected, public).



---
 4. Level

Abstraction → Design Level (conceptual).

Encapsulation → Implementation Level (practical coding).



---

 5. Access

    - Abstraction → Hides only unnecessary details.

     - Encapsulation → Restricts direct access to data and provides controlled access through methods (getters/setters).



---

#5- What are dunder methods in Python?

 - Dunder methods (short for double underscore methods) are special predefined methods in Python whose names begin and end with double underscores.

- They are sometimes called “magic methods” because they allow developers to define how objects of a class should behave with built-in operations like +, ==, len(), print(), for loops, etc.

-   Example: _ _init_ _ , __str_ _ , __len _ _ , __add_ _.


  -  Meaning: When you use something like len(obj), internally Python calls obj._len_().



 - Purpose of Dunder Methods:-



1. Control object creation and initialization (_new, __init_).


2. Define human-readable and unambiguous representations (_str, __repr_).


3. Make objects behave like built-in types (_len, __getitem, __setitem_).


4. Support operator overloading (_add, __sub, __mul_, etc.).


5. Define comparisons and equality checks (_eq, __lt, __gt_).


6. Make custom objects iterable (_iter, __next_).


7. Enable use in context managers (_enter, __exit_).


8. Control attribute access (_getattr, __setattr_).


- Commonly used Dunder methods:-


  i.  _init_ → Object initialization

  ii.  _str_ → String output

  iii.  _len_ → Length

  iv. _add_ → Operator overloading

  v.  _getitem_ → Indexing

  vi.  _iter_ → Iteration

  vii.  _enter/exit_ → Context management

  Example:-

   
 class Book:
    
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages
    def __str__(self):   # For print()
        return f"Book: {self.title}, Pages: {self.pages}"
    def __len__(self):   # For len()
        return self.pages
    def __add__(self, other):  # For +
        return self.pages + other.pages
 Creating objects

 book1 = Book("Python Basics", 250)

 book2 = Book("Advanced Python", 350)

 print(book1)          # Uses __str__ → Book: Python Basics, Pages: 250

 print(len(book1))     # Uses __len__ → 250

 print(book1 + book2)  # Uses __add__ → 60

#6- Explain the concept of inheritance in OOP?

-  Inheritance is a fundamental concept in Object-Oriented Programming (OOPS) that allows one class (child/derived class) to acquire the properties (variables/attributes) and behaviors (methods/functions) of another class (parent/base class).
It promotes code reusability, extensibility, and creates a relationship between classes in a hierarchical manner.


- we use Inheritance:-


  i. To achieve code reusability (write once, use multiple times).

  ii. To implement real-world relationships (e.g., a Dog is a type of Animal).

  iii. To make the code more structured and modular.

  iv.To allow method overriding (child class can redefine parent methods).




 - types of Inheritance:-


1. Single Inheritance → One child class inherits from one parent class.


2. Multiple Inheritance → One child class inherits from multiple parent classes.


3. Multilevel Inheritance → A class is derived from another derived class (grandparent → parent → child).


4. Hierarchical Inheritance → Multiple child classes inherit from a single parent class.


5. Hybrid Inheritance → A combination of two or more types of inheritance.



 Example:-



Parent Class

 Child Class (inherits from Animal)

 Object creation

Here:

#7- What is polymorphism in OOP?

- Polymorphism is a core concept in Object-Oriented Programming that means “many forms”.
It allows objects of different classes to be treated as objects of a common superclass, particularly when they share a common interface or method.

- In simple words: The same method or operator can behave differently depending on the object it is acting on.


- Importance of polymorphism:-

  i. It increases flexibility in code.

  ii. Supports method overriding and method overloading.

  iii. Allows uniform handling of different objects through a common interface.

  iv. Makes code extensible and maintainable.




 - Types of Polymorphism:-

  1- Compile-Time Polymorphism (Static Polymorphism)

- Definition:

    - The method to be executed is decided at compile time.

     - Achieved through method overloading or operator overloading.


-  Characteristics:

    - Occurs within the same class.

     - Method names are same, but parameters differ (number, type, or order).

     - Faster because the method is resolved during compilation.



Example:-

class Calculator:
    
    def add(self, a, b, c=0):
        return a + b + c


calc = Calculator()

print(calc.add(2, 3))      # Output: 5

print(calc.add(2, 3, 4))   # Output: 9


Here, the add method behaves differently depending on the number of arguments.


Operator Overloading Example:

print(5 + 3)       # + operator adds numbers

print("Hi" + "Bye") # + operator concatenates strings

Same operator + works in different ways — this is compile-time polymorphism.



  2️- Run-Time Polymorphism (Dynamic Polymorphism)

- Definition:

   - The method to be executed is decided at runtime, depending on the object type.

    - Achieved through method overriding (child class provides its own implementation of parent method).


- Characteristics:

     - Requires inheritance.

   -  allows flexible behavior for different objects.

     - Slower than compile-time because method resolution happens at runtime.


Example in Python:

class Animal:
    
    def speak(self):
        print("Some sound")

class Dog(Animal):
    
    def speak(self):
        print("Bark")

class Cat(Animal):
    
    def speak(self):
        print("Meow")


animals = [Dog(), Cat(), Animal()]

for a in animals:
    a.speak()

Output:


Bark

Meow

Some sound

The same method speak() behaves differently depending on the object type (Dog, Cat, or Animal).

#8- How is encapsulation achieved in Python?

- Encapsulation is one of the four pillars of OOPS. It is the practice of hiding the internal state of an object and only exposing a controlled interface to access or modify it.

- It protects data from unauthorized access and accidental modification.

- Provides a way to control how data is accessed or modified.


- In simple words: Encapsulation = data hiding + controlled access.






- In Python, encapsulation is achieved using access modifiers:-

1. Public members


2. Protected members


3. Private members




---

  1️ Public Members

Accessible from anywhere, inside or outside the class.

No special symbol is required.


Example:

class Employee:
    
    def _init_(self, name, salary):
        self.name = name      # public
        self.salary = salary  # public


emp = Employee("John", 5000)

print(emp.name)   # Output: John

print(emp.salary) # Output: 5000


---

 2️ Protected Members

Indicated by a single underscore _ prefix.

Should not be accessed directly outside the class, but accessible in subclasses.


Example:

class Employee:
    
    def _init_(self, name, salary):
        self._name = name      # protected
        self._salary = salary  # protected

class Manager(Employee):
    
    def show_salary(self):
        return f"Salary is {self._salary}"

m = Manager("Alice", 8000)

print(m.show_salary())  # Output: Salary is 8000


---

 3️ Private Members

Indicated by a double underscore __ prefix.

Cannot be accessed directly outside the class.

Python uses name mangling to prevent direct access (ClassName_attribute).


Example:

class Employee:
    def _init_(self, name, salary):
        self.__name = name      # private
        self.__salary = salary  # private

    # Getter method
    def get_salary(self):
        return self.__salary

    # Setter method
    def set_salary(self, salary):
        if salary > 0:
            self.__salary = salary
        else:
            print("Invalid salary")


emp = Employee("John", 5000)

print(emp.get_salary())   # Access via getter → Output: 5000

emp.set_salary(6000)      # Modify via setter

print(emp.get_salary())   # Output: 6000

# 10-What are class and static methods in Python?

# - Class Methods

  - Definition:

     - A class method is a method that belongs to the class rather than an instance of the class.

    - It can access and modify class-level attributes (shared by all instances).

    - It is defined using the @classmethod decorator.

     - Its first parameter is cls, which refers to the class itself.


-   Syntax:

class ClassName:
    
    @classmethod
    def method_name(cls, parameters):
        # method body

  - Example:

class Employee:
    company = "ABC Corp"  # class attribute

    @classmethod
    def show_company(cls):
        print(f"Company: {cls.company}")

Employee.show_company()  # Output: Company: ABC Corp

 Changing class attribute using class method

Employee.company = "XYZ Ltd"

Employee.show_company()  # Output: Company: XYZ Ltd

-  Key Points:-

   i. Can access/modify class attributes.

   ii. Cannot directly access instance attributes.

    iii. Can be called using class name or instance.




# Static Methods

 - Definition:

    - A static method is a method that does not access instance or class attributes.

   - It behaves like a regular function inside a class.

    - It is defined using the @staticmethod decorator.


Syntax:

class ClassName:
    
    @staticmethod
    def method_name(parameters):
        # method body

Example:

class Math:
    
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(5, 3))  # Output: 8

  - Key Points:

    i. Cannot access instance (self) or class (cls) attributes.

    ii. Used for utility/helper functions inside a class.

     iii. Can be called using class name or instance.

# 11- What is method overloading in Python?

- Method overloading is the ability to define multiple methods with the same name but different parameters (different number or types of arguments) within the same class.

- Purpose:

  - Provides flexibility to use the same method for different types of inputs.

   - Enhances code readability and usability.





- Method Overloading in Python:-



    - Python does not support traditional method overloading like Java or C++.

   - In Python, if you define multiple methods with the same name, the last one overrides the previous definitions.

- However, method overloading can be simulated using:

1. Default arguments


2. Variable-length arguments (*args and **kwargs)






a) Using Default Arguments

class Calculator:
  
    def add(self, a, b=0, c=0):
        return a + b + c


calc = Calculator()

print(calc.add(5))        # Output: 5

print(calc.add(5, 10))    # Output: 15

print(calc.add(5, 10, 15))# Output: 30

The same method add works with 1, 2, or 3 arguments using default values.





b) Using *args (Variable-length Arguments)

class Calculator:
    
    def add(self, *numbers):
        return sum(numbers)


calc = Calculator()

print(calc.add(5))           # Output: 5

print(calc.add(5, 10))       # Output: 15

print(calc.add(5, 10, 15))   # Output: 30

*numbers collects all arguments as a tuple and allows any number of arguments.

# 13- What is a property decorator in Python?

 - A property decorator in Python is a built-in feature (@property) that allows methods in a class to be accessed like attributes. It is mainly used to define getter, setter, and deleter methods in a clean and Pythonic way, while still providing data encapsulation.






- We use a property decorator to:-

   - Access methods like attributes (without parentheses).

    - Add validation or logic while setting or getting values.

     - Protect private attributes (encapsulation).

    - Make code more readable and maintainable.




1. @property → defines a getter method.


2. @<property_name>.setter → defines a setter method.


3. @<property_name>.deleter → defines a deleter method.






 Example:-


class Student:
    
    def _init_(self, name, marks):
        self._name = name
        self._marks = marks

    # getter
    @property
    def marks(self):
        return self._marks

    # setter
    @marks.setter
    def marks(self, value):
        if value < 0:
            raise ValueError("Marks cannot be negative!")
        self._marks = value

 Usage

s = Student("Alice", 80)

print(s.marks)    # getter -> 80

s.marks = 90      # setter -> updates marks

#15- What is an abstract class in Python?

 - An abstract class in Python is a class that cannot be instantiated directly and is designed to be a blueprint for other classes. It can contain one or more abstract methods (methods declared but not implemented). Subclasses of the abstract class must provide implementations for these abstract methods.



 - Python provides the abc (Abstract Base Class) module. To create an abstract class:

1. Import ABC and abstractmethod from abc.


2. Inherit the class from ABC.


3. Define abstract methods using @abstractmethod.


Example:-


from abc import ABC, abstractmethod


class Animal(ABC):   # Abstract class
    
    @abstractmethod
    def sound(self):
        pass


   class Dog(Animal):   # Subclass must implement sound()
    
    def sound(self):
        return "Bark"

class Cat(Animal):
    
    def sound(self):
        return "Meow"

 Usage

dog = Dog()

cat = Cat()

print(dog.sound())  # Bark

print(cat.sound())  # Meow

 animal = Animal()  # ❌ Error: Can't instantiate abstract class





- We use abstract classes to:

  i. Enforce a common interface for all subclasses.

  ii. Ensure that subclasses must implement certain methods.

  iii. Support abstraction by hiding implementation details and showing only essential behavior.

   Provide base functionality that can be extended by subclasses.

#17- What is the difference between a class variable and an instance variable?

# Class Variable:-

 - A class variable is a variable that is shared by all objects (instances) of a class. It is defined inside the class but outside any method. If the variable is updated at the class level, the change reflects in all instances (unless overridden in a specific object).



# instance variable:-


- An instance variable belongs to a specific object (instance) of the class. It is usually defined inside the constructor (_init_) using self. Each object gets its own copy, and changing it in one object does not affect others.



-  Class Variable vs Instance Variable

   a)  Belongs to:

    Class Variable → Class (shared by all objects)

     Instance Variable → Object (unique to each instance)


   b) Defined in:

      Class Variable → Inside class, outside methods

      Instance Variable → Inside constructor (_init_) using self


   c) Copies:

      Class Variable → One copy, shared by all objects

      Instance Variable → Separate copy for each object


   d) Change effect:

      Class Variable → Change reflects in all objects

      Instance Variable → Change affects only that object

 Example:-



class Student:
    
    school_name = "ABC School"   # class variable

    def _init_(self, name, age):
        self.name = name         # instance variable
        self.age = age           # instance variable

 Create objects

s1 = Student("Alice", 15)

s2 = Student("Bob", 16)

print(s1.school_name)  # ABC School

print(s2.school_name)  # ABC School

 Change class variable


Student.school_name = "XYZ School"

print(s1.school_name)  # XYZ School

print(s2.school_name)  # XYZ School

 Change instance variable

s1.age = 20

print(s1.age)  # 20 (only for s1)

print(s2.age)  # 16 (unaffected)

#18- What is multiple inheritance in Python?

- Definition:-

   - Multiple Inheritance in Python is a feature where a child class can inherit properties and methods from more than one parent class. This allows a single class to combine behaviors from multiple classes.




- uses of Multiple Inheritance:-



  i. To reuse code from multiple classes.

#22- What is the difference between @staticmethod and @classmethod in Python?


# Static Method vs Class Method:-

1. Decorator Used:

    - Static Method → @staticmethod

    - Class Method → @classmethod



2. First Parameter:

    - Static Method → No mandatory first parameter (self or cls not required)

    - Class Method → First parameter is cls (refers to the class)



3. Access:

    - Static Method → Cannot access instance variables or class variables directly

    - Class Method → Can access class variables but not instance variables



4. Purpose:

    - Static Method → Performs a function related to the class but independent of class or instance data

    - Class Method → Works with the class itself and can modify class state



5. Calling:

    - Both can be called using class name or instance, but Static Method doesn’t get automatic class/instance reference.



6. Use Case:

    - Static Method → Utility functions that don’t need class/instance data

    - Class Method → Factory methods or methods that need to modify class-level data





---

Example:

    @staticmethod
    def greet():
        print("Hello from static method")

    @classmethod
    def increment(cls):
        cls.count += 1

Bark

Meow

Some sound

The speak() method is common in all classes.

```python
 # Base class
 class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate
 # Derived class
 class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus
    def calculate_salary(self):
        return super().calculate_salary() + self.bonus
 # Testing
 emp = Employee("vishakha", 60, 30)
 mgr = Manager("komal", 40, 20, 500)
 print("Employee salary:", emp.calculate_salary())
 print("Manager salary:", mgr.calculate_salary())
```
